```python
import tkinter as tk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
question=[
    "hii",
    "hello",
    "how are you",
    "who are you",
    "what can you do"]
answer=[
    "Hii😊",
    "Hello there! How can i assist you today?",
    "I am doing well😊i am hear to chat with you!",
    "I am your ML chatbot hear to help you with Machine Learning🙌",
    "I am your Ai powered assistant!"]
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(question)
def get_response(user_input):
    user_vec=vectorizer.transform([user_input])
    similarity=cosine_similarity(X,user_vec)
    idx=similarity.argmax()
    score=similarity.max()
    if score<0.2:
        return "I didn't understand😥"
    return answer[idx]

root=tk.Tk()
root.title("Keethx ML chatbot🐱‍💻")
root.geometry("400x500")

chat_frame=tk.Frame(root,bg="#7291d6")
chat_frame.pack(padx=10,pady=10,fill=tk.BOTH,expand=True)
entry=tk.Entry(root,font=("Arial",12),bg="#9fb4e0")
entry.pack(padx=10,pady=10,fill=tk.X)
def send_message():
    user_msg=entry.get()
    if user_msg=="":
        return
    user_label = tk.Label(
        chat_frame,
        text="You: " + user_msg,
        bg="#DCF8C6",
        fg="black",
        padx=10,
        pady=5)
    
    
    user_label.pack(anchor="e", pady=5, padx=10)

    # BOT BOX
    bot_reply = get_response(user_msg)
    
    bot_label = tk.Label(
        chat_frame,
        text="Bot: " + bot_reply,
        bg="#EAEAEA",
        fg="black",
        padx=10,
        pady=5)   
    
    
    bot_label.pack(anchor="w", pady=5, padx=10)
    #chat_area.insert(tk.END,"You:  " + user_msg + "\n")
    #bot_reply=get_response(user_msg)
    #chat_area.insert(tk.END,"Bot:  " + bot_reply + "\n\n")
    entry.delete(0,tk.END)
root.bind('<Return>', lambda event: send_message())
button=tk.Button(root,text="Send",bg="#9397db",command=send_message).pack()
root.mainloop()
```
